**Pie Chart**

```r
x <- c(10,20,30,40)
pie(x)
```

```r
#create a vector f pies
x <- c(10,20,30,40)
#Display the pie chat amd start the first pie at 90 degrees
pie(x,init.angle=90)
```

```r
mylabel <- c("Apples","Bananas","Cherries","Dates")
pie(x,label=mylabel,main="Fruits")
```

```r
colors <- c("blue","yellow","green","black")
pie(x,label=mylabel,main="Fruits",col=colors)
#bottom right is the position
legend("bottomright",mylabel,fill=colors)
```

**Bar Plot**


```r
x <- c("A","B","C","D")
y <- c(2,4,6,8)
barplot(y,name.arg=x)
```

**Horizontal Bars**

If you want the bars to be displayed horizontally instead of vertically,use horiz=TRUE.

```r
x <- c("A","B","C","D")
y <- c(2,4,6,8)
barplot(y,names.arg = x, horiz = TRUE,col=colors)
mylabel <- c("A","B","C","D")
legend("bottomright",mylabel,fill=colors)
```

**Histogram**

```r
#Create data for the graph
v <- c(9,13,21,8,36,22,12,41,31,33,19)
#Create histogram
hist(v,xlab = "Weight",col = "yellow", border = "blue")
```

```r
#Create data for the graph
v <-c(9,13,21,8,36,22,12,41,31,33,19)
#Create histogram
hist(v,xlab = "Weight",col = "purple", border = "blue",xlim = c(0,40), ylim = c(0,5),breaks = 5)
```
